### Introdução:

Os quimiostatos são sistemas amplamente utilizados em processos biotecnológicos para o cultivo contínuo de microrganismos sob condições controladas. Nesse tipo de reator, nutrientes são continuamente adicionados ao sistema enquanto uma quantidade de biomassa é removida, mantendo o volume constante. Essa operação permite estudar o crescimento microbiano, o consumo de substratos e a influência de diferentes parâmetros operacionais sobre a dinâmica da cultura.

O modelo matemático de um quimiostato é uma ferramenta importante para compreender e prever o comportamento desses sistemas. Entre os modelos mais utilizados destaca-se a cinética de Monod, que descreve a relação entre a taxa específica de crescimento dos microrganismos e a concentração do substrato disponível. Por meio dessa abordagem, foi possível analisar como fatores como taxa máxima de crescimento, constante de saturação, taxa de diluição e rendimento(biomassa/substrato) afetam o número desses microorganismos e a concentração de nutrientes ao longo do tempo.

Neste trabalho foi desenvolvido um simulador computacional simples utilizando a linguagem Python, com as bibliotecas Tkinter e Matplotlib, para representar o funcionamento de um quimiostato. A interface gráfica permite ao usuário modificar os parâmetros do modelo e acompanhar a evolução do crescimento microbiano e da concentração de substrato por meio de um gráfico dinâmico. Dessa forma, o simulador é uma ferramenta didática para o estudo da linguagem Python e como as suas aplicações englobam outras áreas da ciência.

### Objetivo:

Desenvolver uma simulação simplificada de um quimiostato utilizando a linguagem Python, baseado na cinética de crescimento microbiano de Monod, para analisar a dinâmica da biomassa e do substrato em um cultivo contínuo.

### Resultados:

Os resultados obtidos por meio da simulação demonstraram um comportamento compatível com o esperado para um sistema de quimiostato descrito pela cinética de Monod. Inicialmente, a concentração de biomassa aumentou devido à disponibilidade de substrato no meio, enquanto a concentração de substrato sofreu redução em razão do consumo pelos microrganismos. 

Com o avanço da simulação, observou-se uma tendência à estabilização das variáveis, indicando a aproximação do sistema a um estado estacionário, condição característica de quimiostatos operando em regime contínuo. Além disso, verificou-se a influência dos parâmetros do modelo sobre a dinâmica do sistema, especialmente da taxa de diluição (D), que controla simultaneamente a entrada de nutrientes e a remoção de biomassa. 

Os resultados também evidenciaram que valores elevados de D podem levar à diminuição significativa da biomassa, caracterizando o fenômeno de washout, no qual os microrganismos são removidos do reator mais rapidamente do que conseguem se reproduzir. Dessa forma, a simulação permitiu visualizar de maneira intuitiva as relações entre crescimento microbiano, consumo de substrato e condições operacionais do quimiostato, além de validar a implementação computacional do modelo matemático em Python.

#### Código:

##### Importando as bibliotecas que serão usadas:
1. tkinter: biblioteca que será usada para fazer a interface gráfica
2. matplotlib.figure import Figure:  cria o gráfico
3. matplotlib.backends.backend_tkagg import FigureCanvasTkAgg: permite colocar o gráfico na interface criada

```python
import tkinter as tk
from matplotlib.figure import Figure
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
```

##### Essa célula cria o estado inicial do sistema:
X, S e t são variáveis que representam o estado inicial do sistema. 

O contador será usado posteriormente em atualizar_gráfico() para que o gráfico não seja atualizado milhares de vezes para que a interface não fique lenta.

As listas são usadas para armazenar os resultados para plotar o gráfico e assim o matplotlib.figure consegue desenhar as curvas.

A variável rodando é uma variável booleana que descreve se a simulação está ativa (True) ou não (False).

```python
# Variáveis iniciais

X = 0.5   # biomassa (concentração)
S = 5   # substrato (concentração)
t = 0     # tempo

contador = 0 

    # Listas para gráfico

tempo = []
biomassa = []
substrato = []

rodando = False
```

#####  Essa célula pega os valores digitados pelos usuários:
        
As funções mu_max, Ks, D, Sin e Y pegam o valor digitado pelo usuário em entrada_função.get() e transforma ele em um número real com float.

A função mu(S) implementa a equação de Monod

```python
# Funções

def mu_max():

    '''Retorna o valor digitado pelo usuário da taxa de crescimento máxima'''
    
    return float(entrada_mu.get())


def Ks(): 

    '''Retorna o valor digitado pelo usuário da constante de saturação'''
    
    return float(entrada_Ks.get())
    

def D(): # taxa de diluição: renovação do meio

    '''Retorna o valor digitado pelo usuário da taxa de diluição'''
    
    return float(entrada_D.get())


def Sin():

    '''Retorna o valor digitado pelo usuário da concentração do substrato de entrada'''

    return float(entrada_Sin.get())


def Y(): # redimento (biomassa/substrato)

    '''Retorna o valor digitado pelo usuário do rendimento'''

    return float(entrada_Y.get())


    # Função de Crescimento(Monod):
    
def mu(S): 

    '''Retorna o valor da equaçao de monod'''
    
    return mu_max() * S / (Ks() + S) 
```

##### Essa célula cria uma função para atualizar o gráfico. 
Primeiro ax.clear() limpa o gráfico antigo e sem isso as cuvas seriam desenhadas umas sobre as outras.

As linhas onde ax.plot estão indica a plotagem do gráfico com duas curvas: uma para a biomassa e outra para o substrato.

```python
# Gráfico

def atualizar_grafico():
    
    '''Atualiza o gráfico na tela'''
    
    ax.clear() 
    
    ax.plot(tempo, biomassa,label="Biomassa (X)") 
    ax.plot(tempo, substrato, label="Substrato (S)") 
    ax.set_title("Simulação de Quimiostato")
    ax.set_xlabel("Tempo(h)")
    ax.set_ylabel("Concentração (g/l)")
    
    ax.legend() #adiciona a legenda
    
    canvas.draw() #atualiza o gráfico na tela
```

##### Essa célula cria uma funçao que calcula a evolução do gráfico:
A linha 5 permite alterar as variáveis criadas fora da função. Ocorre também, na linha 7, a verificação se a simulação está ativa ou não, se não estiver rodando a função termina imediatamente. Em dt = 0.01, indica que cada iteração avança 0.01 unidades de tempo.

Em dX e dS ocorre o cálculo da variação da biomassa e do substrato. E logo em seguida foi usado o Método de Euler para o valor calcular o novo valor de X e S, em relação ao atual.

A última linha indica que a função passo() só será executada novamente após 10 ms.

```python
# Simulação

def passo(): 

    '''Calcula a evolução do crescimento e depois atualiza o gráfico'''
    
    global X, S, t, contador 

    if not rodando: # se a simulação estiver pausada, para a execução.
        return

    dt = 0.015 # intervalo de tempo 

    # Equações do quimiostato
    
    crescimento = mu(S)
    
    dX = (crescimento - D()) * X 
    
    dS = D() * (Sin() - S) - (1 / Y()) * crescimento * X 

    # Método de Euler + evita valores negativos
    
    X = max(0, X + dX * dt)
    S = max(0, S + dS * dt)

    t += dt 

    # Salva os dados nas listas para os gráficos
    
    tempo.append(t)
    biomassa.append(X)
    substrato.append(S)

    contador += 1

    # Atualiza o gráfico somente a cada 20 passos:
    
    if contador % 20 == 0:
        atualizar_grafico()

    janela.after(10, passo)
```

##### Essa célula cria os controles para serem usados na interface:

```python
# Controles

    
def iniciar(): 

    '''Inicia a simulação'''
    
    global rodando

    rodando = True
    
    passo()


    
def parar():

    '''Pausa a simulação'''
    
    global rodando
    
    rodando = False
    

    
def resetar():

    '''Reseta as variáveis no gráfico'''
    
    global X, S, t, tempo, biomassa, substrato, contador
    
    X = 0.5 # volta ao estágio inicial
    S = 5
    t = 0

    contador = 0 
    
    tempo = [] 
    biomassa = []
    substrato = []
    
    atualizar_grafico()
```

##### Essa célula cria a inteface:
Na parte da criação dos botões o command = ... , ele executa a função definida anteriormente.
        
A função criar_input() foi criada para evitar repetições no código

```python
# Interface 

janela = tk.Tk() #cria a janela 
janela.title("Simulador de Quimiostato")

area_inputs = tk.Frame(janela) #cria área dos inputs

area_inputs.pack() #sem ele a área existe, mas não aparece

def criar_input(label, valor): 
    
    '''Retorna o campo onde o usuário coloca os valores'''

    tk.Label(area_inputs, text=label).pack() #cria texto do campo
    
    e = tk.Entry(area_inputs) #cria caixa de entrada
    e.insert(0, valor) #coloca valor padrão
    e.pack() #mostra na tela
    
    return e 

# Cria todos os parâmetros do modelo:
    
entrada_mu = criar_input(
    "Taxa de Crescimento Máxima(μmax)", "0.5"
)

entrada_Ks = criar_input(
    "Constante de Saturação (Ks)", "0.1"
)

entrada_D = criar_input(
    "Taxa de diluição (D)", "0.2"
)

entrada_Sin = criar_input(
    "Concentração do substrato de entrada(Sin)", "5"
)

entrada_Y = criar_input(
    "Rendimento (Y)", "0.5"
)

# Botões

frame_botoes = tk.Frame(janela)
frame_botoes.pack()

#cria os botões e liga às funções:

tk.Button(
    frame_botoes,
    text="Iniciar",
    command=iniciar
).pack(side=tk.LEFT)

tk.Button(
    frame_botoes,
    text="Parar",
    command=parar
).pack(side=tk.LEFT)

tk.Button(
    frame_botoes,
    text="Resetar",
    command=resetar
).pack(side=tk.LEFT)

#Aviso ao usuário:
tk.Label(
    area_inputs,
    text="Atenção: se D ≥ μmax, pode ocorrer washout (lavagem da cultura).",
    fg="red"
).pack(pady=5)
```

##### Criação do gráfico e inserção dele na interface.

```python
# Gráfico 

fig = Figure(figsize=(5, 4), dpi=100) #cria a figura e os eixos

ax = fig.add_subplot(111) #1 linha, 1 coluna e gráfico número 1 


# Coloca o gráfico na interface

canvas = FigureCanvasTkAgg(fig, master=janela) 

canvas.get_tk_widget().pack()


# Loop principal ("motor" da interface):

janela.mainloop()
```
